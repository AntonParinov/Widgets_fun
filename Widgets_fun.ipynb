```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

from ipywidgets import interact
%matplotlib widget
```

```python
# 1. Setup - Create the 'Canvas' once
# We give it a clear ID (num=1) so it doesn't get lost


plt.close('all') 
fig = plt.figure(num=1, figsize=(7, 7))
ax = fig.add_subplot(111)

def test_func(x):
    ax.clear() # Clear the old drawing
    
    # Draw the 'Tent' (Triangle)
    ax.plot([0, x, 0], [0, x, 0], marker='o', color='red', linewidth=3)
    
    # Fix the view so it doesn't jump
    ax.set_ylim(0, 20)
    ax.set_xlim(-0, 20)
    ax.set_title(f"Height is: {x}")
    ax.grid(True)
    
    # FORCE the update to the browser
    fig.canvas.draw() 

# 2. Display the initial empty plot
plt.show()

# 3. Start the interaction
interact(test_func, x=(0, 20))
```
